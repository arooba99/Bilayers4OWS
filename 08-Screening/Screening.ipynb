```python
import pandas as pd
%matplotlib inline
```

```python
unlabeled_set = pd.read_csv('unlabeled.csv')
unlabeled_set
```

```python
formula = unlabeled_set['Formula']
prototype = unlabeled_set['prototype']
mismatch = unlabeled_set['Mismatch']
lattice_param = unlabeled_set['avg_lattice_param']
band_alignment = unlabeled_set['band_alignment']

Eb = pd.read_csv('../05-TargetProperties/Unlabeled_Eb.csv')
do = pd.read_csv('../05-TargetProperties/Unlabeled_ILD.csv')

Eg = pd.read_csv('../05-TargetProperties/Unlabeled_Eg.csv')
IE = pd.read_csv('../05-TargetProperties/Unlabeled_IE.csv')
EA = pd.read_csv('../05-TargetProperties/Unlabeled_EA.csv')

wf = pd.read_csv('../05-TargetProperties/Unlabeled_wf.csv')
Ecbm = pd.read_csv('../05-TargetProperties/Unlabeled_Ecbm.csv')
Evbm = pd.read_csv('../05-TargetProperties/Unlabeled_Evbm.csv')
```

```python
pred_df = pd.concat([formula, prototype, mismatch, lattice_param, Eb['Unlabeled_Eb'], do['Unlabeled_ILD'], Eg['Unlabeled_Eg'], band_alignment, IE['Unlabeled_IE'], EA['Unlabeled_EA'], wf['Unlabeled_wf'], Ecbm['Unlabeled_Ecbm'], Evbm['Unlabeled_Evbm']], axis=1)
pred_df
```

```python
hetero_elec = pred_df.loc[pred_df["Unlabeled_Eg"] >=0]
hetero_elec
```

```python
hetero_elec.to_csv('Non_metallic_pred.csv', index=False)
```

```python
hetero_pcs = hetero_elec.loc[hetero_elec["Unlabeled_Eb"] >0]
hetero_pcs
```

```python
hetero_HER = hetero_pcs.loc[hetero_pcs["Unlabeled_Ecbm"] >=-4.44]
hetero_HER
```

```python
hetero_HER.to_csv('HER_pred.csv', index=False)
```

```python
hetero_OER = hetero_pcs.loc[hetero_pcs["Unlabeled_Evbm"] <=-5.67]
hetero_OER
```

```python
hetero_OER.to_csv('OER_pred.csv', index=False)
```

```python
hetero_OWS = hetero_HER.loc[hetero_HER["Unlabeled_Evbm"] <=-5.67]
hetero_OWS
```

```python
hetero_OWS = hetero_OWS.round(2)
hetero_OWS.to_csv('ML-PredictedPhotocatalysts.csv', index=False)
hetero_OWS
```
