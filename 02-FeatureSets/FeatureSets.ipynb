```python
import pandas as pd
```

```python
# Load the trainset CSV file
train_df = pd.read_csv('labeled.csv')
test_df = pd.read_csv('unlabeled.csv')
```

```python
# Split Labeled Dataset into different sets of features
formula_df = train_df.iloc[:, [0]]          # formula
prototype_df = train_df.iloc[:, [1]]
set1_df = train_df.iloc[:, 2:29]            # c2db monolayer features
set2_df = train_df.iloc[:, 29:45]           # band_dos features
a1 = train_df.iloc[:, 45:46]                # lattice parameter
set3_df = train_df.iloc[:, 46:106]          # matminer composition-based features
set4_df = train_df.iloc[:, 106:224]         # magpie composition-based features
set5_df = train_df.iloc[:, 224:248]         # edge-based features
set6_df = train_df.iloc[:, 248:254]         # work function based features
type_df = train_df.iloc[:, 254:259]         # huid and type features

target_df = train_df.iloc[:, 259:]          # DFT-calculated target properties
```

```python
target_df
```

```python
formula_df.to_csv('bilayer_formula.csv', index=False)
a1.to_csv('bilayer_lattice_param.csv', index=False)
target_df.to_csv('../05-TargetProperties/target_properties.csv', index=False)
target_df.to_pickle('../05-TargetProperties/target_properties.pkl')
```

```python
all_feat = pd.concat([set1_df, set2_df, a1, set3_df, set4_df, set5_df, set6_df, type_df], axis=1)
all_feat.to_pickle('all_features.pkl')
all_feat.to_csv('all_features.csv', index=False)
```

```python
train01 = pd.concat([set1_df, set2_df, set5_df, a1], axis=1)
train02 = pd.concat([set1_df, set2_df, set3_df, set4_df, a1], axis=1)
train03 = pd.concat([set1_df, set2_df, set3_df, set5_df, a1], axis=1)               # with Matminer, no wf
train04 = pd.concat([set1_df, set2_df, set3_df, set5_df, set6_df, a1], axis=1)      # with wf
```

```python
train01.to_pickle('train01_features.pkl')
train02.to_pickle('train02_features.pkl')
train03.to_pickle('train03_features.pkl')
train04.to_pickle('train04_features.pkl')
```
